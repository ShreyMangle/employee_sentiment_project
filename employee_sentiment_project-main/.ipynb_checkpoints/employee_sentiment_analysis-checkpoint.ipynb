```python
if not pd.api.types.is_datetime64_any_dtype(df['date']):
    df['date'] = pd.to_datetime(df['date'], errors='coerce')
else:
    df['date'] = df['date']
print('Parsed date -- nulls:', df['date'].isna().sum())
df['date'].dtype
```

```python
# Simple rule-based sentiment labeling
POS_WORDS = set(['love','great','good','happy','supportive','opportunities','growth','enjoy','excellent','satisfied','pleased','awesome'])
NEG_WORDS = set(['unhappy','poor','overworked','underpaid','toxic','stressful','leave','leaving','angry','bad','frustrated','problem'])

def rule_based_sentiment(text):
    txt = str(text).lower()
    pos = sum(1 for w in POS_WORDS if w in txt)
    neg = sum(1 for w in NEG_WORDS if w in txt)
    score = 0.0
    if pos+neg > 0:
        score = (pos - neg) / (pos + neg)
    label = 'Neutral'
    if score > 0.2:
        label = 'Positive'
    elif score < -0.2:
        label = 'Negative'
    return label, score

feedback_col = 'Subject'
if feedback_col is None:
    raise ValueError('No feedback/text column identified in dataset.')
df[['sentiment_label','sentiment_score']] = df[feedback_col].apply(lambda t: rule_based_sentiment(t)).apply(pd.Series)
df[['sentiment_label','sentiment_score']].head()
```

```python
# EDA
import matplotlib.pyplot as plt

df['text_length'] = df['Subject'].astype(str).str.len()
plt.figure(figsize=(8,4))
df['text_length'].hist(bins=30)
plt.title('Feedback text length distribution')
plt.show()

plt.figure(figsize=(6,4))
df['sentiment_label'].value_counts().plot(kind='bar')
plt.title('Sentiment label counts')
plt.show()

if False:
    plt.figure(figsize=(6,4))
    df[''].value_counts().head(10).plot(kind='bar')
    plt.title('Top departments by feedback count')
    plt.show()
```

```python
# Monthly sentiment scoring
map_num = {'Negative': -1, 'Neutral': 0, 'Positive': 1}
df['sentiment_num'] = df['sentiment_label'].map(map_num)

df['month'] = pd.to_datetime(df['date']).dt.to_period('M')
monthly = df.groupby('month')['sentiment_num'].mean().reset_index()
monthly['month'] = monthly['month'].dt.to_timestamp()

plt.figure(figsize=(8,4))
plt.plot(monthly['month'], monthly['sentiment_num'], marker='o')
plt.title('Monthly Mean Sentiment')
plt.xlabel('Month')
plt.ylabel('Mean Sentiment')
plt.grid(True)
plt.show()

# Save monthly CSV
monthly.to_csv('monthly_sentiment.csv', index=False)
monthly.head()
```

```python
# --- Sentiment Analysis using TextBlob ---
# The function below calculates sentiment polarity for each feedback text.
# Polarity score ranges from -1 (negative) to +1 (positive).


def get_sentiment(text):
    """Return polarity score for the given text using TextBlob."""
    return TextBlob(str(text)).sentiment.polarity

df['polarity'] = df['Subject'].apply(get_sentiment);
df.head()
```

```python
# Employee ranking
rank_df = df.groupby('Subject').agg(
    mean_sentiment=('sentiment_num','mean'),
    n_feedback=('sentiment_num','size')
).reset_index()

rank_df['rank_score'] = rank_df['mean_sentiment'] * np.log1p(rank_df['n_feedback'])
rank_df = rank_df.sort_values('rank_score', ascending=False)

print(rank_df.head(15))
rank_df.to_csv('employee_ranking.csv', index=False)
```

```python
# Flight risk identification (heuristic)
recent_cutoff = pd.to_datetime(df['date']).max() - pd.Timedelta(days=90)
recent = df[pd.to_datetime(df['date']) >= recent_cutoff]

agg_recent = recent.groupby('Subject').agg(
    recent_mean=('sentiment_num','mean'),
    recent_count=('sentiment_num','size')
).reset_index()

slopes = []
for emp, g in df.groupby('Subject'):
    tmp = g.sort_values('date')
    if len(tmp) < 3:
        slopes.append((emp, 0.0))
        continue
    xs = (pd.to_datetime(tmp['date']) - pd.to_datetime(tmp['date']).min()).dt.days.values.reshape(-1,1)
    ys = tmp['sentiment_num'].values
    lr = LinearRegression().fit(xs, ys)
    slopes.append((emp, float(lr.coef_[0])))

slope_df = pd.DataFrame(slopes, columns=['Subject','slope'])
merged = agg_recent.merge(slope_df, on='Subject', how='left')
merged['risk_score'] = (-merged['recent_mean']) + (-merged['slope'] * 30)
merged['flight_risk'] = merged['risk_score'] > 0.3

print(merged.sort_values('risk_score', ascending=False).head(20))
merged.to_csv('flight_risk.csv', index=False)
```

```python
# Linear regression on monthly mean sentiment
monthly = pd.read_csv('monthly_sentiment.csv')
monthly['month'] = pd.to_datetime(monthly['month'])
monthly['month_idx'] = (monthly['month'] - monthly['month'].min()).dt.days
X = monthly['month_idx'].values.reshape(-1,1)
y = monthly['sentiment_num'].values
model = LinearRegression().fit(X, y)
monthly['pred'] = model.predict(X)

plt.figure(figsize=(8,4))
plt.plot(monthly['month'], y, marker='o', label='Actual')
plt.plot(monthly['month'], monthly['pred'], marker='x', linestyle='--', label='Pred')
plt.legend()
plt.title('Monthly Sentiment Trend (Linear Regression)')
plt.show()

print('MSE:', ((y - monthly['pred'])**2).mean())
```
